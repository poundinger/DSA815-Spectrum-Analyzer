```python
inst = rm.open_resource('USB0::0x1AB1::0x0960::DSA8A224501273::INSTR')
print(inst.query("*IDN?"))
```

```python
print(inst.query(":SENSe:FREQuency:CENTer?"))
print(inst.query(":SENS:FREQ:CENT?"))
```
